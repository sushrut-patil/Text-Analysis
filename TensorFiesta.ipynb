#Translate all text to English

```python
from googletrans import Translator
import time
from tqdm import tqdm

translator = Translator()

# Use tqdm with pandas apply method
tqdm.pandas()

# Define a function to translate text
def translate_text(text):
    time.sleep(0.4)
    try:
        if (translator.detect(text).lang != 'en'):
            translated = translator.translate(text, dest="en").text
            return translated
        else:
            return text
    except Exception as e:
        # If there's an error during translation, return the original text
        print(f"Translation failed for text: {text}. Error: {e}")
        return text

# Update the REVIEW_CONTENT column
df["REVIEW_CONTENT"] = df["REVIEW_CONTENT"].progress_apply(translate_text)
```

```python
import re

# Remove all special characters from the REVIEW_CONTENT column
df['REVIEW_CONTENT'] = df['REVIEW_CONTENT'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))

# Remove all non-English words from the REVIEW_CONTENT column
df['REVIEW_CONTENT'] = df['REVIEW_CONTENT'].apply(lambda x: ' '.join([word for word in x.split() if word.isalpha()]))

df['REVIEW_CONTENT'] = df['REVIEW_CONTENT'].apply(lambda x: ' '.join([word.lower() for word in x.split() if word.isalpha()]))
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def filter_similar_sentences(sentences):
    filtered_sentences = []

    vectorizer = TfidfVectorizer()

    tfidf_matrix = vectorizer.fit_transform(sentences)

    cosine_similarities = cosine_similarity(tfidf_matrix, tfidf_matrix)

    for i, sentence in enumerate(sentences):
        is_unique = True
        for j, similarity_score in enumerate(cosine_similarities[i]):
            if i != j:

                if similarity_score > 0.6:
                    is_unique = False
                    break
        if is_unique:
            filtered_sentences.append(sentence)

    return filtered_sentences
```

```python
from tqdm import tqdm
data["REVIEW_LIST"] = data["REVIEW_LIST"].astype(object)
count = 0
for index,row in tqdm(data.iterrows()):
    sku = row['SKU']
    sentences = df[df['SKU']== sku]
    sentences = sentences["REVIEW_CONTENT"]
    List = filter_similar_sentences(sentences)
    count += len(List)
    data.at[index, "REVIEW_LIST"] = List

print(len(df) - count)
```

```python
import pandas as pd
from tqdm import tqdm
from transformers import T5TokenizerFast, T5ForConditionalGeneration


# Initialize the T5 model and tokenizer
tokenizer = T5TokenizerFast.from_pretrained("t5-large", max_length=4048)
model = T5ForConditionalGeneration.from_pretrained("t5-large")

# Create a new DataFrame for summaries
new_df = pd.DataFrame(columns=['SKU', 'SUMMARY'])

# Split long strings into smaller chunks and summarize
for idx, row in tqdm(data.iterrows()):
    original_string = "\n".join(data['REVIEW_LIST'].iloc[idx])
    chunk_size = 4000
    chunks = [original_string[i:i + chunk_size] for i in range(0, len(original_string), chunk_size)]

    chunk_summaries = []
    for chunk in chunks:
        if (len(chunk) <= 100):
          chunk_summaries.append(chunk)
        else:
          input_ids = tokenizer.encode(chunk, return_tensors="pt", max_length=4048, truncation=True)
          summary = model.generate(input_ids, max_length=100, num_beams=4, early_stopping=True)[0]
          chunk_summaries.append(tokenizer.decode(summary))

    # Concatenate chunk summaries
    final_summary = ' '.join(chunk_summaries)

    # Add to the new DataFrame
    new_df.loc[idx] = [row['SKU'], final_summary]

# Inspect the new DataFrame
print(new_df.head())
```
